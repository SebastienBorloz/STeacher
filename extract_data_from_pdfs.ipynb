```python
import pymupdf
from PIL import Image
from LLM_wrap import LLM, img_to_b64
from utilities import read_file
```

```python
model = "gpt-4.1-mini"
gpt = LLM(model)
```

```python
direc = "R4"
doc = pymupdf.open("./pdfs/exos R4/Série 1 - Exercice.pdf")
n = 0
```

```python
# loop
while n < len(doc):
	pix = doc[n].get_pixmap().pil_image()
	pix.save(f"./images/from_pdfs/{direc}/page-{n+1}.png")
	b64_image = img_to_b64(f"./images/from_pdfs/{direc}/page-{n+1}.png")
	prompt = read_file("./prompts/latex_gen.md")
	resp, use = gpt.ask_LLM_txt_and_img(prompt, b64_image, temperature=0)
	with open(f"./images/from_pdfs/{direc}/page-{n+1}.txt", "w") as f:
		f.write(resp)
	n += 1
```
